```python
from pathlib import Path
import jsonref

schema_path = Path("../src/bdp_lib/schemas/bdp.schema.json").absolute()


with schema_path.open() as file:
    schema = jsonref.load(file, base_uri=schema_path.as_uri())
```

```python
def write_schema_description(schema, level=0):
    print(f"{'  ' * level} - {schema['title']}: {schema['description']}")
    if 'properties' in schema:
        for key, value in schema['properties'].items():
            write_schema_description(value, level + 1)
    elif "items" in schema:
        schema = schema["items"]
        if 'properties' in schema:
            for key, value in schema['properties'].items():
                write_schema_description(value, level + 1)

write_schema_description(schema)
```
