```python
from pathlib import Path
import jsonref

"""schema_path1 = Path("../src/bdp_lib/schemas/toolbox.schema.json").absolute()
schema_path2 = Path("../src/bdp_lib/schemas/model.schema.json").absolute()


with schema_path1.open() as file:
    schema1 = jsonref.load(file, base_uri=schema_path1.as_uri())

with schema_path2.open() as file:
    schema2 = jsonref.load(file, base_uri=schema_path2.as_uri())
"""
schema_path = Path("../src/bdp_lib/schemas/bdp.schema.json").absolute()

with schema_path.open() as file:
    schema = jsonref.load(file, base_uri=schema_path.as_uri())

def format_json_schema_draft07(schema, indent=0):
    """Formats a Draft-07 JSON Schema into a readable type definition."""
    spacing = "  " * indent
    
    if "type" in schema:
        if schema["type"] == "object":
            properties = schema.get("properties", {})
            required = schema.get("required", [])
            formatted_props = "\n".join(
                f"{spacing}  {key}: {format_json_schema_draft07(value, indent + 1)}{' (required)' if key in required else ''}"
                for key, value in properties.items()
            )
            return f"object {{\n{formatted_props}\n{spacing}}}"
        elif schema["type"] == "array":
            items = schema.get("items", {})
            return f"array[{format_json_schema_draft07(items, indent)}]"
        else:
            return schema["type"]
    
    elif "$ref" in schema:
        return f"ref({schema['$ref']})"
    
    elif "enum" in schema:
        return f"enum({', '.join(map(str, schema['enum']))})"
    
    elif "anyOf" in schema:
        return " | ".join(format_json_schema_draft07(sub_schema, indent) for sub_schema in schema["anyOf"])
    
    elif "oneOf" in schema:
        return "one of (" + ", ".join(format_json_schema_draft07(sub_schema, indent) for sub_schema in schema["oneOf"]) + ")"
    
    elif "allOf" in schema:
        return "all of (" + ", ".join(format_json_schema_draft07(sub_schema, indent) for sub_schema in schema["allOf"]) + ")"
    
    return "unknown"

print(format_json_schema_draft07(schema))
"""print("Toolbox schema:")
print(format_json_schema_draft07(schema1))
print()
print()
print("Model schema:")
print(format_json_schema_draft07(schema2))"""
```
